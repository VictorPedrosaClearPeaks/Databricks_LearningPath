```python
%sql
CREATE SCHEMA IF NOT EXISTS `gold_layer`;

DROP TABLE IF EXISTS gold_layer.gld_dim_entity;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_entity(
integration_id LONG,
entity_key LONG,
entity_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_vertical;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_vertical(
integration_id LONG,
vertical_key LONG,
vertical_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_service;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_service(
integration_id LONG,
service_key LONG,
service_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_request_status;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_request_status(
integration_id LONG,
status_key LONG,
status_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_sla_status;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_sla_status(
integration_id LONG,
sla_status_key LONG,
sla_status_name STRING,
sla_status_order INT,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_stage;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_stage(
integration_id LONG,
stage_key LONG,
stage_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_service_request;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_service_request(
integration_id STRING,
universal_request_pk STRING,
request_number STRING,
assigned_to STRING,
assignment_group STRING,
status_key LONG,
sla_status_key LONG,
updated_date_ts TIMESTAMP,
closed_date_ts TIMESTAMP,
complaints STRING,
business_percentage FLOAT,
sla_start_date_ts TIMESTAMP,
sla_end_date_ts TIMESTAMP,
sla_planned_end_date_ts TIMESTAMP,
has_breached BOOLEAN,
score INT,
percentage_satisfaction FLOAT,
remaining_days INT,
breached_sla_days INT,
opened_by STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN,
data_source STRING
);

DROP TABLE IF EXISTS gold_layer.gld_dim_calendar;

CREATE TABLE IF NOT EXISTS gold_layer.gld_dim_calendar (
integration_id INT,
calendar_key INT,
date DATE,
year INT,
month INT,
month_name STRING,
month_name_short STRING,
quarter_name STRING,
quarter_number INT,
day INT,
weekday STRING,
weekday_number INT,
week_number INT,
is_current_week_flag BOOLEAN,
is_current_month_flag BOOLEAN,
is_current_quarter_flag BOOLEAN,
is_current_year_flag BOOLEAN,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
start_ts TIMESTAMP,
end_ts TIMESTAMP,
current_flag BOOLEAN
);
```
