```python
%sql
CREATE SCHEMA IF NOT EXISTS `silver_layer`;

DROP TABLE IF EXISTS silver_layer.slvr_entity_hist_id;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_entity_hist_id(
entity_key LONG GENERATED ALWAYS AS IDENTITY,
entity_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_entity;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_entity(
integration_id LONG,
entity_key LONG,
entity_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_vertical_hist_id;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_vertical_hist_id(
vertical_key LONG GENERATED ALWAYS AS IDENTITY,
vertical_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_vertical;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_vertical(
integration_id LONG,
vertical_key LONG,
vertical_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_service_hist_id;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_service_hist_id(
service_key LONG GENERATED ALWAYS AS IDENTITY,
service_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_service;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_service(
integration_id LONG,
service_key LONG,
service_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_request_status_hist_id;


DROP TABLE IF EXISTS silver_layer.slvr_request_status;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_request_status(
integration_id LONG,
status_key LONG,
status_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_sla_status_hist_id;

DROP TABLE IF EXISTS silver_layer.slvr_sla_status;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_sla_status(
integration_id LONG,
sla_status_key LONG,
sla_status_name STRING,
sla_status_order INT,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);


DROP TABLE IF EXISTS silver_layer.slvr_stage_hist_id;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_stage_hist_id(
stage_key LONG GENERATED ALWAYS AS IDENTITY,
stage_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);

DROP TABLE IF EXISTS silver_layer.slvr_stage;

CREATE TABLE IF NOT EXISTS silver_layer.slvr_stage(
integration_id LONG,
stage_key LONG,
stage_name STRING,
created_at_ts TIMESTAMP,
modified_at_ts TIMESTAMP,
data_source STRING
);
```
