```python
%sql


MERGE INTO gold_layer.gld_dim_calendar target
USING (
SELECT
-1 AS integration_id,
-1 AS calendar_key,
TO_TIMESTAMP('1900-01-01') AS date,
0 AS year,
0 AS month,
'Unknown' AS month_name,
'Unknown' AS month_name_short,
'Unknown' AS quarter_name,
0 AS quarter_number,
0 AS day,
'Unknown' AS weekday,
0 AS weekday_number,
0 AS week_number,
FALSE AS is_current_week_flag,
FALSE AS is_current_month_flag,
FALSE AS is_current_quarter_flag,
FALSE AS is_current_year_flag,
TO_TIMESTAMP('1900-01-01') AS created_at_ts,
TO_TIMESTAMP('1900-01-01') AS modified_at_ts,
TO_TIMESTAMP('1900-01-01') AS start_ts,
TO_TIMESTAMP('1900-01-01') AS end_ts,
FALSE AS current_flag
UNION ALL  
SELECT
integration_id,
calendar_key,    
date,
year,
month,
month_name,
month_name_short,
quarter_name,
quarter_number,
day,
weekday,
weekday_number,
week_number,
is_current_week_flag,
is_current_month_flag,
is_current_quarter_flag,
is_current_year_flag,
created_at_ts,
modified_at_ts,
current_timestamp() AS start_ts,
NULL AS end_ts,
TRUE AS current_flag
FROM silver_layer.slvr_calendar
where calendar_key != -1
) SOURCE
ON target.integration_id = source.integration_id
WHEN MATCHED THEN
UPDATE SET 
is_current_week_flag = source.is_current_week_flag,
is_current_month_flag = source.is_current_month_flag,
is_current_quarter_flag = source.is_current_quarter_flag,
is_current_year_flag = source.is_current_year_flag,
modified_at_ts = current_timestamp()
WHEN NOT MATCHED
  THEN INSERT (
integration_id,
calendar_key,    
date,
year,
month,
month_name,
month_name_short,
quarter_name,
quarter_number,
day,
weekday,
weekday_number,
week_number,
is_current_week_flag,
is_current_month_flag,
is_current_quarter_flag,
is_current_year_flag,
created_at_ts,
modified_at_ts,
start_ts,
end_ts,
current_flag)
  VALUES(
source.integration_id,
source.calendar_key,
source.date,
source.year,
source.month,
source.month_name,
source.month_name_short,
source.quarter_name,
source.quarter_number,
source.day,
source.weekday,
source.weekday_number,
source.week_number,
source.is_current_week_flag,
source.is_current_month_flag,
source.is_current_quarter_flag,
source.is_current_year_flag,
source.created_at_ts,
source.modified_at_ts,
source.start_ts,
source.end_ts,
source.current_flag)
;

SELECT * 
FROM gold_layer.gld_dim_calendar
ORDER BY date asc;
```
